Cleaning Data

Only USA races, 50km or 50miles, 2020

Step 1, show 50miles or 50 km

```python
#df[((df['Event distance/length'] == '50km') | (df['Event distance/length'] == '50mi')) & (df['Year of event']==2020)]
df1=df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event']==2020) & (df['Event name'].str.endswith(r'(USA)'))]
```

```python
df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0)=='USA']
```

```python
df2=df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event']==2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0)=='USA')]
```

```python
df1[~df2.isin(df1)].dropna(how='all')
```

```python
df2.head(10)
```

```python
df2.shape
```

Remove USA from event name

```python
df2['Event name'] =df2['Event name'].str.split('(').str.get(0)
```

Get the age of athletes

```python
df2['Athlete Age']= 2020-df2['Athlete year of birth']
```

Remove 'h' from athlete performance

```python
df2['Athlete performance']=df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head(10)
```

Drop column: Athelete Club, Athlete Country, Athlete Year of Birth, Athlete Age Category

```python
df2=df2.drop(['Athlete club','Athlete country','Athlete year of birth','Athlete age category'],axis=1)
```

```python
df2.head()
```

Clean null values

```python
df2.isna().sum()
```

```python
df2[df2['Athlete Age'].isna()==1]
```

```python
df2=df2.dropna()
```

```python
df2.shape
```

Check for duplicate values

```python
df2[df2.duplicated()==True]
```

Reset Index

```python
df2.reset_index(drop=True)
```

Fix types

```python
df2.dtypes
```

```python
df2['Athlete Age'] = df2['Athlete Age'].astype(int)
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

Rename columns

```python
df2=df2.rename(columns={'Year of event': 'year',
                    'Event dates':'race_date',
                    'Event name':'race_name',
                    'Event distance/length':'race_length',
                    'Event number of finishers':'race_number_of_finishers',
                    'Athlete performance':'athlete_performance',
                    'Athlete gender':'gender',
                    'Athlete average speed':'athlete_avg_speed',
                    'Athlete ID': 'athlete_id',
                    'Athlete Age': 'athlete_age'
})
```

Reorder columns

```python
df3=df2[['race_date','race_name','race_length','race_number_of_finishers','athlete_id','gender','athlete_age','athlete_performance','athlete_avg_speed']]
```
